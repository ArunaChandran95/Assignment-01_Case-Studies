# Insights from the dataset: 

# 1. Differenciating the frequency of each flower sample:

```python
data.columns
```

```python
plt.figure(figsize=(5,5))
plt.hist(data['Classification'], color='green',rwidth=0.4)
plt.title('Distribution of different Samples',fontsize=10)
plt.xlabel('Count of Each Sample')   
plt.ylabel('Frequency')
plt.xticks()
plt.yticks()           
           
```

# 2. Comparison of Sepel length & Petal Length:

```python
data.columns
```

```python
plt.figure(figsize=(5,5))
data [['SL','PL']].plot.kde()
plt.title('Comparison of Sepel length & Petal Length' ,fontsize=10)
plt.xlabel('Sepel Length SL')   
plt.ylabel('Petal Length PL')
plt.xticks()
plt.yticks()  
```

# 3. Differenciating the petal width of all variety iris:

```python
plt.figure(figsize=(5,5))
plt.scatter(data ['PW'],data['Classification'],s=12,c= 'violet',marker='^')
plt.title('Comparison of Petal Width with all variety iris flowers',fontsize=10)
plt.xlabel('Petal Width PW',fontsize=10)   
plt.ylabel('Classification',fontsize=10)
```

# Combined comparison of SW and PW with all variety iris:

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.scatter(data ['SW'],data['Classification'],s=12,c= 'red',marker='^')
plt.title('Comparison of Sepal Width with all variety iris flowers',fontsize=10)
plt.xlabel('Sepel Width SW',fontsize=10)   
plt.ylabel('Classification',fontsize=10)

plt.subplot(2,1,2)
plt.scatter(data ['PW'],data['Classification'],s=12,c= 'blue',marker='^')
plt.title('Comparison of Petal Width with all variety iris flowers',fontsize=10)
plt.xlabel('Petal Width PW',fontsize=10)   
plt.ylabel('Classification',fontsize=10)

plt.tight_layout()
```
